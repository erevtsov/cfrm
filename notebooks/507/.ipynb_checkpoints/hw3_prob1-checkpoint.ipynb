```python
import numpy as np
import pandas as pd
```

```python
n = 20
```

```python
df = pd.DataFrame(index=list(range(1, n+1)), columns=['hire', 'skip'])
df.index.name = 'candidate'

def decision(s):
    if s == n + 1:
        return 1
    else:
        no_hire = 0.78 * decision(s+1)
        hire = 0.22 * ((1/11) * 25 + (10/11) + 5)
        df.loc[s, 'hire'] = hire
        df.loc[s, 'skip'] = no_hire
        return no_hire + hire

decision(1)
df['decision'] = np.where(df['hire'] > df['skip'], 'hire', 'skip')
df
```
