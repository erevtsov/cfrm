### Problem 1

```python
import numpy as np
import pandas as pd

n = 20
df = pd.DataFrame(index=list(range(1, n+1)), columns=['hire', 'skip'])
df.index.name = 'candidate'

def decision(s):
    if s == n + 1:
        return 1
    else:
        no_hire = 0.78 * decision(s+1)
        hire = 0.22 * ((1/11) * 25 + (10/11) + 5)
        df.loc[s, 'hire'] = hire
        df.loc[s, 'skip'] = no_hire
        return no_hire + hire

decision(1)
df['decision'] = np.where(df['hire'] > df['skip'], 'hire', 'skip')
df
```

### Problem 3

```python
import numpy as np
import pandas as pd
import cvxpy as cp

# define inputs - delivery cost and construction cost
delivery_cost = np.array([
    [2, 1, 8, 5, 7, 1, 4, 6, 5, 9],
    [4, 9, 4, 3, 9, 4, 4, 2, 7, 2],
    [5, 8, 7, 6, 7, 9, 3, 10, 4, 5],
    [3, 5, 7, 9, 7, 6, 6, 5, 4, 5],
    [8, 7, 3, 6, 8, 5, 7, 4, 8, 7],
    [7, 6, 4, 5, 3, 7, 2, 9, 7, 5],
])
construction_cost = np.array([10, 12, 7, 8, 6, 13])
n_stores = delivery_cost.shape[1]

# construct variables for open warehouses
# and which warehouses supply which store
W = cp.Variable(shape=construction_cost.shape, boolean=True)
S = cp.Variable(shape=delivery_cost.shape, boolean=True)

# create constraints
constraints = []
# each store has exactly one supplier
constraints.append(
    S.T @ np.ones(S.shape[0]) == 1
)
# if store is being supplied by a warehouse
# that warehouse should be open
constraints.append(
    S @ np.ones(S.shape[1]) <= n_stores*W
)

# objective is to minimize sum of costs
obj = cp.Minimize(
    (W @ construction_cost) # construction costs
    + cp.sum(cp.multiply(delivery_cost, S)) # delivery costs
)

prob = cp.Problem(
    objective=obj,
    constraints=constraints,
)
# use mixed-integer GLPK solver
prob.solve(solver='GLPK_MI')

print(f'Warehouses that should be opened:{[str(x + 1) for x in np.where(W.value == 1)]}')
```
