# Revtsov Final Project Part 2

The analysis began with the use of weekly returns and provided weights to run performance attribution, resulting in a stream of weekly returns for each asset category, analysis dimension, and attribution bucket. This stream of weekly excess returns was then subjected to t-tests to determine whether the mean of the excess return vector is statistically different from zero. I calculated the bounds of the 95% confidence interval, and if zero was not part of this interval, it indicated a statistically significant outperformance in that dimension.

#### Performance Attribution Review
##### 60/30/10 Multi-Asset
The overall performance for the 60/30/10 Multi-Asset class was approximately 3 basis points higher on a weekly basis compared to the benchmark. The underweighting of Alternatives was a significant contributor throughout the time period; however, it was not statistically significant according to the t-test, as the 95% confidence interval included zero. The Stock Selection bucket was the largest contributor to outperformance and was statistically significant.

##### 50/45/5 Fund
In the 50/45/5 Fund, there was statistically significant outperformance in the Stock Selection dimension. Positive contributions came from selection within all three categories. Allocations contributed positively overall, with Alternatives leading the charge.

##### Alternatives
Within the Alternatives asset class, the underweighting of Listed Real Estate contributed positively, while other allocation decisions had negative contributions to return. Interestingly, the allocation decision to real estate led to outperformance by 260 basis points over the period, but this was not statistically significant due to significant outliers during the initial COVID period. Selection was statistically significant across all three categories, with Commodities and Real Estate providing positive impacts, while Listed Infrastructure detracted from performance.

##### Bonds
In the Bonds category, overall allocation decisions and within-bond categories led to underperformance. However, selection within long-term treasuries significantly contributed positively, resulting in a positive overall selection bucket despite detracting performances within Investment Grade (IG) and short-term treasuries. Allocation among different credit quality buckets did not significantly impact performance. Nonetheless, selection within treasuries contributed positively, while selection within IG detracted, yielding a net positive overall contribution. For the duration dimension, allocations to both long and short treasuries detracted from performance, but selection within long-term treasuries was the largest contributor to active outperformance.

##### Stocks
Allocation within stocks contributed positively overall, with selection being the key driver. While small-cap selection detracted slightly, both non-US and US large-cap selections led to significant outperformance. Allocation to US stocks contributed positively, whereas non-US allocations detracted, consistent with the broad US stock market performance in recent years.

##### Overall Analysis
Overall, the selection teams are contributing positively to performance across most dimensions of analysis. Selection within Listed Infrastructure and Investment Grade Corporate Bonds are the largest detractors. Selection has driven a significant portion of the outperformance of actual funds versus their benchmarks. Allocation results were mixed, with the interaction effect having the least impact.

## Code

```python
#### Read in Data
import scipy
import numpy as np
import pandas as pd

idx = pd.IndexSlice

bm_rts = pd.read_excel('FinalProjectData.xlsx', sheet_name='Benchmark Returns')
bm_rts.date = pd.to_datetime(bm_rts.date).dt.date
bm_rts.set_index('date', inplace=True)
bm_rts.columns.name = 'BMTicker'

# SHY return is risk free rate
rfr = bm_rts['SHY']

act_rts = pd.read_excel('FinalProjectData.xlsx', sheet_name='Active Returns')
act_rts.date = pd.to_datetime(act_rts.date).dt.date
act_rts.set_index('date', inplace=True)
act_rts.columns.name = 'ActTicker'
last_obs = act_rts.index[-1]

#### Table of Asset Level Data

asset_data = pd.DataFrame( 
    data=[
        ['Stocks', 'U.S. Large Cap', 'IWB', 'AF_IWB', 0.55, 0.58],
        ['Stocks', 'U.S. Small Cap', 'IWM', 'AF_IWM', 0.06, 0.07],
        ['Stocks', 'Non-U.S. Markets', 'IXUS', 'AF_IXUS', 0.39, 0.35],
        ['Bonds', 'Short Treasuries', 'SHV', 'AF_SHV', 0.2, 0.18],
        ['Bonds', 'Long Treasuries', 'TLT', 'AF_TLT', 0.3, 0.33],
        ['Bonds', 'Investment Grade Corporate Bonds', 'IGSB', 'AF_IGSB', 0.5, 0.49],
        ['Alternatives', 'Listed Real Estate', 'XLRE', 'AF_XLRE', 0.34, 0.4],
        ['Alternatives', 'Commodities', 'GSG', 'AF_GSG', 0.33, 0.3],
        ['Alternatives', 'Listed Infrastructure', 'IGF', 'AF_IGF', 0.33, 0.3],
    ],
    columns=[
        'Category', 'AssetClass', 'BMTicker', 'ActTicker', 'BMWeight', 'ActWeight',
    ]
)

#### Function to Calculate Attribution

def calc_attribution(subset, name, dimension, scale=1e4, round_digit=4):
    # if either of the BM or Actual weight vectors don't add up to 1, rescale them.
    subset.loc[:, 'BMWeight'] = subset.BMWeight / subset.groupby('date').BMWeight.sum()
    subset.loc[:, 'ActWeight'] = subset.ActWeight / subset.groupby('date').ActWeight.sum()
    
    # calculate the 3 quadrants
    q1 = subset.BMWeight * subset.BMReturn
    q2 = subset.ActWeight * subset.BMReturn
    q3 = subset.BMWeight * subset.ActReturn
    q4 = subset.ActWeight * subset.ActReturn
    
    # calculate attribution and return
    attribution = (q2 - q1).rename('Allocation').to_frame().join(
        (q3 - q1).rename('Selection')
    ).join(
        (q4 - q2 - q3 + q1).rename('Other')
    )
    attribution.index.names = ['date', 'Category']
    attribution['Name'] = name
    attribution['Dimension'] = dimension
    return attribution.reset_index().set_index(['date', 'Name', 'Dimension', 'Category']).mul(scale).round(round_digit)

#### Merge in Returns

all_data = asset_data.merge(
    bm_rts.stack().rename('BMReturn').reset_index(), on='BMTicker', how='inner'
).merge(
    act_rts.stack().rename('ActReturn').reset_index(), on=['date', 'ActTicker'], how='inner'
)
all_data.loc[:, 'BMCtr'] = all_data.BMWeight * all_data.BMReturn
all_data.loc[:, 'ActCtr'] = all_data.ActWeight * all_data.ActReturn

#### Calculate Return of Multi-Asset funds

multi_asset_rts = all_data.groupby(['date', 'Category'])[['BMCtr', 'ActCtr']].sum()
multi_asset_rts.columns = ['BMReturn', 'ActReturn']

#### 60/30/10 Fund
multi_asset_fund1 = pd.DataFrame(
    data=[
        ['Stocks', .6, .59],
        ['Bonds', .3, .26],
        ['Alternatives', .1, .15]
    ],
    columns=['Category', 'BMWeight', 'ActWeight'],
).set_index('Category')
multi_asset_fund1 = multi_asset_fund1.join(multi_asset_rts)

#### 50/45/5 Fund
multi_asset_fund2 = pd.DataFrame(
    data=[
        ['Stocks', .5, .51],
        ['Bonds', .45, .42],
        ['Alternatives', .05, .07]
    ],
    columns=['Category', 'BMWeight', 'ActWeight'],
).set_index('Category')
multi_asset_fund2 = multi_asset_fund2.join(multi_asset_rts)

#### Run Attribution

frames = []
## Multi-Asset Analysis

#### 60/30/10 Fund
frames.append(calc_attribution(multi_asset_fund1, name='60/30/10 Fund', dimension='Allocation'))

#### 50/45/5 Fund
frames.append(calc_attribution(multi_asset_fund2, name='50/45/5 Fund', dimension='Allocation'))

## Asset Class Analysis

### Stocks

#### Allocation

predicate = 'Category == "Stocks"'
subset = all_data.query(predicate).copy().set_index(['date', 'AssetClass'])
frames.append(calc_attribution(subset, name='Stocks', dimension='Allocation'))

#### Capitalization

predicate = '(Category == "Stocks") & (AssetClass != "Non-U.S. Markets")'
subset = all_data.query(predicate).copy()

subset.loc[subset.AssetClass.str.contains('Large'), 'Cap'] = 'Large'
subset.loc[subset.AssetClass.str.contains('Small'), 'Cap'] = 'Small'

subset = subset.set_index(['date', 'Cap'])
frames.append(calc_attribution(subset, name='Stocks', dimension='Cap'))

#### Country

predicate = 'Category == "Stocks"'
subset = all_data.query(predicate).copy()

subset['Country'] = 'US'
subset.loc[subset.AssetClass.str.contains('Non-U.S.'), 'Country'] = 'Non-US'

subset = subset.groupby(['date', 'Country'])[['BMWeight', 'ActWeight', 'BMCtr', 'ActCtr']].sum().rename(columns={'BMCtr': 'BMReturn', 'ActCtr': 'ActReturn'})
frames.append(calc_attribution(subset, name='Stocks', dimension='Country'))

### Bonds

#### Allocation

predicate = 'Category == "Bonds"'
subset = all_data.query(predicate).copy().set_index(['date', 'AssetClass'])
frames.append(calc_attribution(subset, name='Bonds', dimension='Allocation'))

#### Duration

predicate = '(Category == "Bonds") & (AssetClass != "Investment Grade Corporate Bonds")'
subset = all_data.query(predicate).copy()

subset.loc[subset.AssetClass.str.contains('Long'), 'Duration'] = 'Long'
subset.loc[subset.AssetClass.str.contains('Short'), 'Duration'] = 'Short'

subset = subset.reset_index().set_index(['date', 'Duration'])
frames.append(calc_attribution(subset, name='Bonds', dimension='Duration'))

#### Credit Quality

predicate = 'Category == "Bonds"'
subset = all_data.query(predicate).copy()

subset.loc[subset.AssetClass.str.contains('Treasuries'), 'Credit Quality'] = 'Treasuries'
subset.loc[subset.AssetClass.str.contains('Investment Grade'), 'Credit Quality'] = 'IG'

subset = subset.groupby(['date', 'Credit Quality'])[['BMWeight', 'ActWeight', 'BMCtr', 'ActCtr']].sum().rename(columns={'BMCtr': 'BMReturn', 'ActCtr': 'ActReturn'})
frames.append(calc_attribution(subset, name='Bonds', dimension='Credit Quality'))

### Alternatives

#### Allocation

predicate = 'Category == "Alternatives"'
subset = all_data.query(predicate).copy().set_index(['date', 'AssetClass'])
frames.append(calc_attribution(subset, name='Alternatives', dimension='Allocation'))

attrib_all = pd.concat(frames)

### Outperformance

fund_rts_ma = pd.concat([
    multi_asset_fund1.groupby('date')[['BMReturn', 'ActReturn']].sum().reset_index().assign(Fund='60/30/10'),
    multi_asset_fund2.groupby('date')[['BMReturn', 'ActReturn']].sum().reset_index().assign(Fund='50/45/5')
]).set_index(['date', 'Fund'])

fund_rts_ac = all_data.groupby(['date', 'Category'])[['BMCtr', 'ActCtr']].sum()
fund_rts_ac.columns = ['BMReturn', 'ActReturn']
fund_rts_ac.index.names = ['date', 'Fund']

fund_rts_subac = all_data.set_index(['date', 'AssetClass'])[['BMReturn', 'ActReturn']]
fund_rts_subac.index.names = ['date', 'Fund']

all_fund_rts = pd.concat([fund_rts_ma, fund_rts_ac, fund_rts_subac])

all_fund_rts = all_fund_rts[['ActReturn', 'BMReturn']] 

all_fund_rts.loc[:, 'ExcessReturn'] = all_fund_rts['ActReturn'] - all_fund_rts['BMReturn']

outperf_stats = all_fund_rts.groupby('Fund')['ExcessReturn'].mean().rename('AvgExcessReturn').to_frame().join(
    all_fund_rts.groupby('Fund')['ExcessReturn'].std().rename('TrackingError')
).mul(1e4)

### Attribution Statistical Significance

dims = attrib_all.reset_index()[['Name', 'Dimension', 'Category']].drop_duplicates().reset_index(drop=True)
pvalue_all = pd.DataFrame(index=pd.MultiIndex.from_frame(dims), columns=attrib_all.columns).sort_index()
tstat_all = pd.DataFrame(index=pd.MultiIndex.from_frame(dims), columns=attrib_all.columns).sort_index()
ci_all = pd.DataFrame(index=pd.MultiIndex.from_frame(dims), columns=attrib_all.columns).sort_index()
ci_mask = pd.DataFrame(index=pd.MultiIndex.from_frame(dims), columns=attrib_all.columns, data=True).sort_index()

for ix, row in dims.iterrows():
    for col in tstat_all.columns:
        ix = idx[:, row['Name'], row['Dimension'], row['Category']]
        xs = attrib_all.loc[ix, col]
        res = scipy.stats.ttest_1samp(
            xs,
            popmean=0,
            # np.random.normal(loc=0., scale=xs.std(), size=xs.shape[0])
        )
        ci = res.confidence_interval(confidence_level=0.95)
        tstat_all.loc[idx[row['Name'], row['Dimension'], row['Category']], col] = res.statistic
        pvalue_all.loc[idx[row['Name'], row['Dimension'], row['Category']], col] = res.pvalue
        ci_all.loc[idx[row['Name'], row['Dimension'], row['Category']], col] = f'({round(ci.low, 4)}, {round(ci.high, 4)})'
        if ci.low < 0 and ci.high > 0:
            ci_mask.loc[idx[row['Name'], row['Dimension'], row['Category']], col] = False

# Define the threshold
threshold = 0.005

def highlight_pvalue(val):
    return 'background-color: lightgreen' if val else ''
```
