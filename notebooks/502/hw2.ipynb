## Problem 1

```r
itb <- getSymbols("ITB", from="2010-01-01", to="2024-01-01", auto.assign=FALSE)
itb.prices <- itb$ITB.Adjusted
```

### Part a

```r
itb.returns <- na.omit(Return.calculate(itb.prices, method = "log"))
```

##### mean

```r
(mu <- mean(itb.returns))
```

##### variance

```r
var(itb.returns)[1, 1]
```

##### std dev

```r
(sig <- sqrt(var(itb.returns)[1, 1]))
```

### Part b

```r
itb.density <- density(itb.returns)
options(repr.plot.width = 12, repr.plot.height = 8)
hist(itb.returns, xlab="Daily Log Returns", main="", probability=TRUE, breaks=131)
lines(itb.density, col="red", lwd=2)
x <- seq(min(itb.returns), max(itb.returns), len=length(itb.returns))
lines(x, dnorm(x, mean=mu, sd=sig), col="blue", lwd=2)
legend("topleft", legend = c("KDE", "Normal PDF"), col = c("red", "blue"), lty = c(1, 1))
```

### Part c

```r
sample = coredata(itb.returns)
```

```r
qqPlot(sample, distribution='norm', envelope=TRUE, id=FALSE)
```

```r
qqPlot(sample, distribution='t', df=4, envelope=TRUE, id=FALSE)
```

The sample return distribution has fat tails. This is shown in the qq plot comparison vs normal distribution as concavity on the left tail and convexity on the right tail. It is confirmed by comparison to the T distribution as it looks like a good fit.

## Problem 3

### Part a

##### bias calculation
$$
    bias[\hat{\theta_n}] = \mathbb{E}[\hat{\theta_n}] - \theta$$

$$
    \hat{\theta_1} = 2 \hat{\mu} = \frac{2}{n} \sum{X_i}
$$

$$
    \hat{\theta} = 2 \hat{\mu} = \mathbb{E}[X] = 2\mu
$$

$$
    \mathbb{E}[\hat{\theta_1}] = \mathbb{E}[\frac{2}{n} \sum{X_i}] = \frac{2}{n}\mathbb{E}[\sum{X_i}] = \frac{2}{n} n \mu = 2 \mu 
$$

$$
    bias[\hat{\theta_1}] = \mathbb{E}[\hat{\theta_1}] - \theta = 2\mu - 2\mu = 0
$$


##### variance calculation
$$
    Var[U(a,b)] = \frac{1}{12}(b-a)^2
$$

$$
    Var[\hat{\theta_1}] = Var[\frac{2}{n} \sum{X_i}] = \frac{4}{n^2} \sum{Var[X_i]} = \frac{4}{n^2} n \frac{\theta^2}{12} = \frac{\theta^2}{3n}$$

$$
    se[\hat{\theta_1}] = \sqrt{Var[\hat{\theta_1}]} = \frac{\theta}{\sqrt{3n}}
$$

### Part b

$$
    \hat{\theta_2} = a\{X_1, X_2, ..., X_n\}=\{aX_1, aX_2, ..., aX_n\}
$$

$$
    F_{\hat{\theta_2}}(x) = P(aX_1 < x)...P(aX_n < x) = P(X_1 < \frac{x}{a})...P(X_n < \frac{x}{a}) = F_x(\frac{x}{a})^n
$$

### Part c

$$
    \mathbb{E}[max\{X_n\}] = \frac{n}{n+1}\quad for\quad X \textasciitilde  U(0,1)
$$

$$
    \mathbb{E}[\hat{\theta_2}] = \mathbb{E}[a*max\{X_n\}] = a\mathbb{E}[max\{X_n\}] = a \theta \frac{n}{n+1}
$$

$$
    bias[\hat{\theta_n}] = \mathbb{E}[\hat{\theta_n}] - \theta = a \theta \frac{n}{n+1} = \theta (a \frac{n}{n+1} - 1)
$$

Value of a for an unbiased estimator is $$a=\frac{n+1}{n}$$

### Part d

$$
    Var[max\{X_n\}] = \frac{n}{(n+1)^2(n+2)}\quad for\quad X \textasciitilde U(0,1)
$$

$$
    Var[\hat{\theta_2}] = Var[a*max\{X_n\}] = a^2 \frac{\theta^2 n}{(n+1)^2(n+2)}
$$

$$
    se[\hat{\theta_2}] = \sqrt{Var[\hat{\theta_2}]} = \sqrt{\frac{(n+1)^2}{n^2} \frac{\theta n}{(n+1)^2(n+2)}} = \sqrt{\frac{\theta}{n(n+2)}}
$$

### Part e

$$
    mse[\hat{\theta_1}] = bias[\hat{\theta_1}]^2 + se[\hat{\theta_1}]^2 = 0 + \frac{\theta^2}{3n}
$$

$$
    mse[\hat{\theta_2}] = bias[\hat{\theta_2}]^2 + se[\hat{\theta_2}]^2 = 0 + \frac{\theta}{n(n+2)}
$$

If theta and n are both greater one, theta 2 has the smaller mse due to smaller numberator and larger denominator, making it the better estimator due to lower mse.
