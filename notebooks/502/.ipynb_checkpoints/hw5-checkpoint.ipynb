## Problem 1

```r
load("big_mac_data.Rdata")
attach(mac)
```

```r
var.names <- names(mac)
var.names <- var.names[var.names != 'BigMac']
```

### Part a

```r
best.subsets <- regsubsets(x=BigMac~Bread+WorkHrs+VacDays+BusFare+Service+TeachSal+TeachTax+EngSal+EngTax,data=mac, nvmax=9)
(b <- summary(best.subsets))
```

```r
par(mfrow=c(1, 3))
plot(1:9, b$adjr2, type="b", main="Adjusted Rˆ2", xlab="p", ylab="Rˆ2")
plot(1:9, b$bic, type="b", main="BIC", xlab="p", ylab="BIC")
plot(1:9, b$cp, type="b", main="Mallows' C_p", xlab="p", ylab="C_p")
```

All 3 approaches are in agreement that a 5-factor model is best; it should include BusFare, Service, TeachSal, TeachTax, EngSal as the explanatory variables.

### Part b

```r
lm_model <- lm(BigMac ~ ., data = mac[var.names])
best_model <- stepAIC(lm_model, direction = "both", trace = FALSE)

summary(best_model)
```

AIC agrees with results in Part a; 5 factor model consisting of BusFare, Service, TeachSal, TeachTax, and EngSal factors is best.

### Part e

```r
n = length(BigMac)
p = length(mac) - 1
thresh <- qf(0.5, df1=p + 1, df2=n - p - 1)

r <- rstudent(best_model)
h <- hatvalues(best_model)
d <- r^2 * h / ((p + 1) * (1 - h))

colors <- rep("black", n); colors[which.max(d)] <- "red"

par(mfrow=c(1, 3))
# residual plot
plot(r, xlab="i", ylab="Studentized Residual", ylim=c(-11, 11), main="BigMac", col=colors)
abline(h=c(-3, 3), col="red")
# leverage plot
plot(h, xlab="i", ylab="Leverage", ylim=c(0, 1), main="BigMac", col=colors)
abline(h=2 * 2 / n, col="red")
# cook's distance plot
plot(d, xlab="i", ylab="Cook's Distance", ylim=c(0, max(1, max(d) * 1.05)), main="BigMac", col=colors)
abline(h=thresh, col="red")
```

The point highlighted red appears to belong to Mexico City. It is the most influential based on Cook's distance. This point has the largest studentized residual from all the observations, but it still lies within the 3 standard deviation threshold; I wouldn't deem it an outright outlier but still make note of the size of the residual and examine it further by looking at leverasge and overall influence. In terms of leverage, there are 3 data points with higher values and a few with leverage about equal; the leverage is around 0.3 while the mean leverage is 0.13, so this point still has high leverage. It's highly influential as its' Cook's distance is more than 3x the next largest observation.

## Problem 2

### Part a

```r
n <- 10000
x <- seq(0, 1, len=n)
eps <- rnorm(n, mean=2, sd=x)
eps <- eps * sign(rnorm(n))
y <- 0 + 0 * x + eps
fit <- lm(formula = y ~ x)
```

```r
car::qqPlot(fit$residuals, distribution='norm', id=FALSE, envelope=TRUE)
```

```r
par(mfrow=c(1, 3))
# Plot versus x
plot(x, rstandard(fit), xlab="x", ylab="r_i")
smoother <- loess(rstandard(fit) ~ x)
lines(x[order(x)], fitted(smoother)[order(x)], col="red")
# Plot versus Yhat
y.hat <- fit$fitted.values
plot(y.hat, rstandard(fit), xlab="Y hat", ylab="r_i")
smoother <- loess(rstandard(fit) ~ y.hat)
lines(y.hat[order(y.hat)], fitted(smoother)[order(y.hat)], col="red")

hist(fit$residuals, breaks = 40, main = "Histogram of Residuals", xlab = "Residuals")
```

The loess curve is flat so the redisual expectations appear to be constant; it's clear that the variance of the residuals is not constant as the dispersion descreases as Y_hat increases. The residuals are not normally distributed based on the QQ plot. We knew that the variance of the residuals would not be constant when creating the distribution (standard deviation increases with each sample); this is seen in the diagnostics since there's a clear trend in the variance of residuals.

### Part b

```r
n <- 10000
x <- seq(0, 1, len=n)
f.x <- 100 * (x <= 0.2) + 100 * (x >= 0.8)
eps <- rnorm(n)
y <- f.x + eps
fit <- lm(formula = y ~ x)
```

```r
car::qqPlot(fit$residuals, distribution='norm', id=FALSE, envelope=TRUE)
```

```r
par(mfrow=c(1, 3))
# Plot versus x
plot(x, rstandard(fit), xlab="x", ylab="r_i")
smoother <- loess(rstandard(fit) ~ x)
lines(x[order(x)], fitted(smoother)[order(x)], col="red")
# Plot versus Yhat
y.hat <- fit$fitted.values
plot(y.hat, rstandard(fit), xlab="Y hat", ylab="r_i")
smoother <- loess(rstandard(fit) ~ y.hat)
lines(y.hat[order(y.hat)], fitted(smoother)[order(y.hat)], col="red")

hist(fit$residuals, breaks = 40, main = "Histogram of Residuals", xlab = "Residuals")
```

The expectations of residuals are not constant since there's a clear hyperbolic shape to the loess curve. The issue of non-linear residual expectations make it difficult to assess variance of residuals. However, it does seem like the variance is constant as the residuals are in 3 clear "blocks" that have similar heights. The residuals are not normally distributed according to the QQ plot. Even though the "eps" variable is normally distributed the actual Y variable is not linear in terms on X and therefore the resulting regression residuals are not going to be normally distributed.
