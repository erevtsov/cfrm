## Problem 1

### Part a
To prove that Y_t is white noise we need to prove that the expectation is zero and that autocorrelation is 1 when h=0 and 0 othersize.
$$
    \mathbb{E}[Y_t] = \mathbb{E}[Z_t Z_{t-1}] = \mathbb{E}[Z_t]\mathbb{E}[Z_{t-1}]=0 
$$

$$
    \mathbb{E}[Z_t^2] = \mathbb{V}[Z_t] + \mathbb{E}[Z_t]^2 = 1+0 = 1
$$

$$
    \mathbb{V}[Y_t] = \mathbb{V}[Z_t Z_{t-1}] = \mathbb{E}[(Z_t Z_{t-1})^2] - \mathbb{E}[Z_t Z_{t-1}]^2 = \mathbb{E}[(Z_t Z_{t-1})^2] = \mathbb{E}[Z_t^2]\mathbb{E}[Z_{t-1}^2] = 1*1=1
$$

$$
\begin{equation}
  \rho(h) = 
  \begin{cases}
  1 & h=0 \\
  0 & h\neq0
  \end{cases}
\end{equation}
$$

$$
    \rho(0) = \frac{{\mathbb{V}[Y_t]}}{{\mathbb{V}[Y_t]}}=1
$$

$$
    \rho(h) = \frac{\gamma(h)}{\gamma(0)} = \frac{\mathbb{C}[Y_t, Y_{t+h}]}{\mathbb{V}[Y_t]} = \mathbb{C}[Y_t, Y_{t+h}]
$$

$$
    \mathbb{C}[Y_t, Y_{t+h}] = \mathbb{E}[Y_t Y_{t+h}] - \mathbb{E}[Y_t]\mathbb{E}[Y_{t+h}] = \mathbb{E}[Y_t Y_{t+h}] = \mathbb{E}[(Z_t Z_{t-1})(Z_{t+h} Z_{t+h-1})]
$$
$$
    \mathbb{C}[Y_t, Y_{t+h}] = \mathbb{E}[Z_t]\mathbb{E}[(Z_{t-1})(Z_{t+h} Z_{t+h-1})] = 0
$$
$$
    Z_t\text{ is independent from others and expecation is zero}
$$

### Part b

$$
    \mathbb{C}[Y_t^2, Y_{t-1}^2] = \mathbb{E}[Y_t^2 Y_{t-1}^2] - \mathbb{E}[Y_t^2] \mathbb{E}[Y_{t-1}^2]
$$
$$
    \mathbb{C}[Y_t^2, Y_{t-1}^2] = \mathbb{E}[Y_t^2 Y_{t-1}^2] - (\mathbb{V}[Y_t]+\mathbb{E}[Y_t]^2)(\mathbb{V}[Y_{t-1}]+\mathbb{E}[Y_{t-1}]^2)
$$
$$
    \mathbb{C}[Y_t^2, Y_{t-1}^2] = \mathbb{E}[Y_t^2 Y_{t-1}^2] - 1
$$
$$
    \text{For }{Y_t}\text{ to be an i.i.d. sequence the above would equal to zero.}
$$
$$
    \mathbb{E}[Y_t^2 Y_{t-1}^2] = \mathbb{E}[(Z_t Z_{t-1})^2 (Z_{t-1} Z_{t-2})^2] = \mathbb{E}[Z_t^2]\mathbb{E}[Z_{t-1}^4]\mathbb{E}[Z_{t-2}^2] = \mathbb{E}[Z_{t-1}^4]
$$
$$
    \mathbb{E}[Z_{t-1}^4] = \mathbb{V}[Z_{t-1}^2] - \mathbb{E}[Z_{t-1}^2]^2 = 3\sigma^4 = 3
$$
$$
    \mathbb{C}[Y_t^2, Y_{t-1}^2] \text{ is not zero.}
$$

### Part a
$$
    X_t - \phi_1 X_{t-1} - \phi_2 X_{t-2} = Z_t
$$
$$
    \mathbb{E}[X_t] - \phi_1 \mathbb{E}[X_{t-1}] - \phi_2 \mathbb{E}[X_{t-2}] = \mathbb{E}[Z_t]
$$
$$
    \mathbb{E}[X_t] - \phi_1 \mathbb{E}[X_{t-1}] - \phi_2 \mathbb{E}[X_{t-2}] = 0
$$
$$
    \text{Since process is stationary } \mathbb{E}[X_t] = \mathbb{E}[X_{t-1}] = \mathbb{E}[X_{t-2}]
$$
$$
    \mathbb{E}[X_t](1 - \phi_1 - \phi_2) = 0
$$
$$
    \text{If } \phi_1 + \phi_2 \neq 1 \text{, then } \mathbb{E}[X_t]=0
$$

### Part e
$$
    se[\hat{\mu_n}] = \frac{\sigma}{\sqrt{n}}
$$
$$
    \gamma(0)=(1+\theta)\sigma^2
$$
$$
    \gamma(1)=\theta\sigma^2
$$
$$
    \gamma(0)+2\gamma(1) = \sigma^2(1+3\theta)
$$
$$
    \sqrt{\gamma(0)+2\gamma(1)} = \sigma \sqrt{(1+3\theta)}
$$
$$
    \sqrt{\frac{1}{n}\gamma(0)+2\gamma(1)} = \frac{\sigma}{\sqrt{n}} \sqrt{(1+3\theta)}
$$

$$
    \sqrt{\frac{1}{n}\gamma(0)+2\gamma(1)} \approx \frac{\sigma}{\sqrt{n}}
$$
$$
    se[\hat{\mu_n}] \approx \sqrt{\frac{1}{n}\gamma(0)+2\gamma(1)}
$$
