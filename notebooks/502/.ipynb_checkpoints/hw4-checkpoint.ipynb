#### Creating a function to run the linear regression. Used throuhgout the homework.

```r
# inputs are the response variablale, data frame of explanatory variables,
# and the desired column names (don't include Intercept)
lin_reg <- function(Y_in, X_in, col_names) {
    Y <- as.matrix(Y_in)
    n <- length(Y_in)
    p <- length(col_names)
    colnames(Y) <- NULL
    X <- cbind(rep(1, n), as.matrix(X_in))
    colnames(X) <- c('Intercept', col_names)

    # calculate beta vector
    # I'm using R version 4.1.3 and it doesn't have "inv" function
    # so using "solve" in its place
    beta_hat <- solve(t(X) %*% X) %*% t(X) %*% Y

    # calculate epsilon var using hat matrix
    H <- X %*% solve(t(X) %*% X) %*% t(X)
    eps_hat <- Y - H %*% Y

    # calculate variance of beta
    eps_var <- as.numeric(t(eps_hat) %*% eps_hat / (n-p-1))
    # extract the diagonal, we're not looking for the covariances right now
    beta_var <- diag(eps_var * solve(t(X) %*% X))

    # confidence interval
    alpha <- 0.05 # could expose this as an input param
    talpha <- qt(1 - alpha / 2, df=n-p-1)
    ci.lower <- beta_hat - talpha * sqrt(beta_var)
    ci.upper <- beta_hat + talpha * sqrt(beta_var)

    # hypothesis test for beta=0
    t_value <- (beta_hat - 0) / sqrt(beta_var)
    p_value <- 2 * pt(-abs(t_value), df=n-p-1)

    # aggregate results into tables
    res <- cbind(beta_hat, sqrt(beta_var), t_value, p_value)
    colnames(res) <- c('Estimate', 'Std Error', 't-value', 'p-value')
    ci.table = cbind(ci.lower, ci.upper)
    colnames(ci.table) = c('lower', 'upper')

    # print results of the regression and confidence internval
    cat('\nRegression Summary\n')
    print(res)
    cat('\nConfidence Interval\n')    
    print(ci.table)
}
```

### Part b

### Part e

* GDP and T-Bill variables have a positive correlation of 0.3 so the positive non-zero coefficient makes sense in the first model.
* adding the Inflation variable doesn't change things much; this variable is uncorrelated with the response variable T-Bill so I wouldn't expect a statistically significant coefficient here.
* Unemployment has a strong negative correlation with T-Bill; in absolute terms, the correlation between T-Bill and Unemployment is stronger than that of T-Bill and GPD. Adding this to our model produces a negative coefficient and decreases the GDP variable significance because Unemployment has better explanatory power.

This shows us that feature selection is extermely important; we may think that a factor drives returns from initial results, but that may be only because we didn't include the correct combination in the model.

### Part a

$$
    Y_i = \beta_0 + \sum{\beta_k X_{i,k}} + \epsilon_i
$$ 
$$
    Y = X\beta + \hat{\epsilon}
$$
$$
    \hat{Y} = X\hat{\beta}
$$
If we can show that epsilon is orthogonal to X that would also show that the expectation of epsilon is zero because X contains a vector of ones and epsilon is orgthonal to every column of X.
$$
    \mathbb{E}[\hat{\epsilon}] = \frac{1}{n}\sum\hat{\epsilon_i} = \hat{\epsilon}^T\mathbb{1}
$$ 
And the first column of X is ones.
$$
    H=X(X^TX)^{-1}X^T
$$
$$
    HX=X(X^TX)^{-1}X^TX=X
$$

$$
    \hat{\epsilon} = Y-\hat{Y} = Y-HY = (I-H)Y
$$
$$
    \hat{\epsilon}^TX = ((I-H)Y)^TX=Y^T(I-H)X=Y^T(X-HX)=0
$$
The above equation is zero because X=HX. We proved that epsilon is orthogonal to a vector of ones (the first column of X is all ones and epsilon is orthogonal to each column). Therefore:
$$
    \hat{\epsilon}^T\mathbb{1} = \mathbb{E}[\hat{\epsilon}] = \frac{1}{n}\sum\hat{\epsilon_i} = 0
$$

Show that covariance of epsilon and betas is zero
$$
    Cov(Y, \hat{\beta}) = Cov(Y, (X^TX)^{-1}X^TY)=Cov(Y,Y)*((X^TX)^{-1}X^T)^T=\sigma_{\epsilon}^2*(X(X^TX)^{-1}

$$
    Cov(\hat{Y}, \hat{\beta})=Cov(X\hat{\beta}, \hat{\beta})=X*Cov(\hat{\beta})=X(\sigma_{\epsilon}^2(X^TX)^{-1})=\sigma_{\epsilon}^2X(X^TX)^{-1}
$$

$$
    Cov(\hat{\epsilon}, \hat{\beta})=Cov(Y,\hat{\beta}) - Cov(\hat{Y}, \hat{\beta})=\sigma_{\epsilon}^2X(X^TX)^{-1}-\sigma_{\epsilon}^2X(X^TX)^{-1}=0
