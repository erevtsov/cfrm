```python
import numpy as np
import pandas as pd
```

## Question 1

```python
def exp_u(g, n=1000):
    n_v = np.array(range(1, n)).astype(float)
    return np.sum((2**(n_v**(1-g)-1)) / (2**(n_v+1-g)))
```

```python
g_v = np.append(np.array(range(1, 100)), (np.array(range(101, 400)))) / 100
```

```python
pi = np.vectorize(exp_u)(g_v)
```

```python
ax = pd.Series(index=g_v, data=pi).plot(kind='line', title='Certainty Equivalent')
ax.set_ylabel('Certainty Equivaltnt (pi)')
ax.set_xlabel('Gamma')
```
