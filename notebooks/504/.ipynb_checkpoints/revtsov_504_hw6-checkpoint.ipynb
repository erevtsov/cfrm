### 6.1c

```python
M = 1
T = 1
N = 100
#
S0 = 1
B0 = 1
mu = 0.1
sigma = 0.4
r = 0.03
p = -2
#
dt = T/N

np.random.seed(42)  # Use any integer as the seed value
```

```python
# normally distributed random variable with mu=0

### STOCK ###
log_S = np.zeros([N+1, M])
log_dS = np.zeros([N+1, M])
# start at S0
log_S[0, :] = np.log(S0)
for j in range(M):
    for i in range(1, N+1):
        z = np.random.normal(loc=0, scale=1)
        log_dS[i, j] = (mu - 1/2 * (sigma**2))*dt + sigma * (dt**(1/2)) * z
        log_S[i, j] = log_S[i-1, j] + log_dS[i, j]

### BOND ###
B = np.zeros(N+1)
dB = np.zeros(N+1)
B[0] = B0
for i in range(1, N+1):
    dB[i] = r*dt*B[i-1]
    B[i] = B[i-1] + dB[i]
```

```python
S = np.exp(log_S)
dS = S[1:, :] - S[:-1, :]
dS = np.append([0] * M, dS)
```

```python
pd.DataFrame({'S': S.mean(axis=1), 'B': B}).plot(kind='line')
```

### 6.1e

```python
X = np.zeros([N+1, M])
delta = np.zeros([N+1, M])
V = np.zeros([N+1, M])

X[0, :] = np.exp(-1 * r * T) * S[0, :] ** p
V[0, :] = np.exp(-1 * r * T) * S[0, :] ** p
delta[0, :] = p * np.exp(-1 * r * T) * S[0, :] ** (p-1)

for j in range(M):
    for i in range(1, N+1): # 0 thru N-1
        t = dt*(i)
        # print(f'{i}: {t}')
        
        # print(S[i, j])
        V[i, j] = np.exp(-1 * r * (T-t)) * (S[i, j] ** p)
        # 
        delta[i, j] = p * np.exp(-1 * r * (T-t)) * (S[i, j] ** (p-1))
        X[i, j] = X[i-1, j] + delta[i-1, j] * dS[i] + (X[i-1, j] - delta[i-1, j] * S[i-1, j]) * (1/B[i-1]) * dB[i]
```

```python
pd.DataFrame({'V': V.mean(axis=1), 'X': X.mean(axis=1)}).plot(kind='line')
```

```python
V[-1, :] == S[-1, :] ** p
```

```python
pd.Series(X[-1, :] - V[-1, :]).plot(kind='hist', bins=50)
```
